```python
documents = [
    "Artificial intelligence is transforming the way we work and live. Machine learning models can now perform complex tasks with high accuracy.",
    "Climate change is one of the biggest challenges facing humanity. Rising temperatures are causing extreme weather events around the world.",
    "Quantum computing promises to revolutionize computing power. Quantum computers can solve certain problems exponentially faster than classical computers.",
    "The human brain contains approximately 86 billion neurons. These neurons communicate through electrical and chemical signals to process information.",
    "Renewable energy sources like solar and wind power are becoming more affordable. These clean energy technologies are essential for a sustainable future."
]
```

```python
def preprocess_text(text):
    """Convert to lowercase and remove punctuation"""
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

preprocessed_docs = [preprocess_text(doc) for doc in documents]

print("Original Documents:")
for i, doc in enumerate(documents, 1):
    print(f"{i}. {doc}\n")

print("\nPreprocessed Documents:")
for i, doc in enumerate(preprocessed_docs, 1):
    print(f"{i}. {doc}\n")
```

```python
print("Loading embedding model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!\n")

print("Creating embeddings...")
embeddings = model.encode(preprocessed_docs)
print(f"Created {len(embeddings)} embeddings of dimension {embeddings[0].shape[0]}\n")

print("Initializing ChromaDB...")
client = chromadb.Client()
```

```python
collection = client.create_collection(name="documents")

collection.add(
    embeddings=embeddings.tolist(),
    documents=documents,
    ids=[f"doc_{i}" for i in range(len(documents))]
)
print("Documents stored in ChromaDB successfully!\n")
```

```python
def query_documents(query_text, n_results=1):
    """Query the document collection and return most relevant document"""
    # Preprocess query
    preprocessed_query = preprocess_text(query_text)

    # Embed query
    query_embedding = model.encode([preprocessed_query])

    # Query ChromaDB
    results = collection.query(
        query_embeddings=query_embedding.tolist(),
        n_results=n_results
    )

    return results
```

```python
test_queries = [
    "Tell me about AI and machine learning",
    "What is happening with the climate?",
    "How does the brain work?",
    "What are the benefits of solar energy?"
]

print("="*70)
print("TESTING RETRIEVAL SYSTEM")
print("="*70)

for query in test_queries:
    print(f"\nQuery: '{query}'")
    print("-" * 70)

    results = query_documents(query)

    # Extract the most relevant document
    relevant_doc = results['documents'][0][0]
    distance = results['distances'][0][0]

    print(f"Most Relevant Document:")
    print(f"{relevant_doc}")
    print(f"\nCosine Distance: {distance:.4f}")
    print(f"Similarity Score: {1 - distance:.4f}")
    print("=" * 70)
```

```python
print("\n\nTry your own query!")
print("Enter a question or type 'exit' to quit\n")

while True:
    user_query = input("Your query: ").strip()

    if user_query.lower() == 'exit':
        print("Goodbye!")
        break

    if not user_query:
        continue

    print("\nSearching...")
    results = query_documents(user_query)

    relevant_doc = results['documents'][0][0]
    distance = results['distances'][0][0]

    print(f"\nMost Relevant Document:")
    print(f"{relevant_doc}")
    print(f"\nSimilarity Score: {1 - distance:.4f}\n")
    print("-" * 70)
```
